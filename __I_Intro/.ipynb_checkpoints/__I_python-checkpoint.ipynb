## 2.Aritmetika ir kintamieji

Viena iš paprasčiausių užduočių, kurią galite paprašyti kompiuterio, yra atspausdinti kokį nors pranešimą.
Python programoje mes prašome kompiuterio atspausdinti mums pranešimą, parašydami `print()` komandą ir įterpdami pranešimą į skliaustus kabutese. Žemiau prašome kompiuterio atspausdinti pranešimą -, „mano pirmoji programa“.

```python
print("mano pirmoji programa")
```

Kodas yra celės viduje (kodo langelis), o kompiuterio atsakymas (vadinamas kodo išvestimi) rodomas po langeliu. Kaip matote, kompiuteris atspausdino pranešimą, kurio mes tikėjomės. Nors mes naudosimės Jupyter sistemos sąsają rašyti, testuoti python programas lygiai taip pat tą galima padaryti naudojant komandinę eilutė iškvietus python interpretatorių su komanda `python`  arba sukurti failą <b>manoprograma.py</b> ir per komandinę eilutę įvedus sekančia komandą: `python manoprograma.py` gausime tą patį rezultatą.

Taip pat galime atspausdinti kai kurių aritmetinių operacijų (pvz., sudėties, atimties, daugybos ar padalijimo) reikšmes.
Pavyzdžiui, kitame kodo langelyje kompiuteris pridės 2 prie 2 ir tada išspausdins rezultatą. Atminkite, kad skirtingai nei tada, kai tiesiog spausdinome tekstą, nenaudojame jokių kabučių.

```python
print(2 + 2)
```

Taip pat galime atlikti atimties operacija.

```python
print(5 - 5)
```

Iš tikrųjų galite atlikti daugybę skaičiavimų naudodami python! Žemiau esančioje lentelėje rasite keletą pavyzdžių.

| Operacija  | Simbolis | Pavyzdys |   |
|          :-|       :- |       :- |:- |
  Sudėtis    | +        | 2 + 2    ||
  Atimtis    | -        | 3 - 1    ||
  Daugyba    | *        | 5 * 5    ||
  Dalyba     | /        | 9 / 3    ||
  Eksponentė | **       | 3 ** 3   ||


Ilgų operacijų tvarką galite valdyti skliausteliuose.Python vadovaujasi PEMDAS taisykle, kai nustato operacijų tvarką.<br>
<code>**PEMDAS taisyklė**</code><br>
P- Parentheses, E- Exponents, M- Multiplication, D- Division, A- Addition, and S- Subtraction.

```python
print(((1 + 3) * (9 - 2) / 2) ** 2)
```

Iki šiol naudojome kodą skaičiavimui ir rezultatui atspausdinti, o rezultatas niekur neišsaugomas. Tačiau galbūt reikės išsaugoti rezultatą, kad programa galėtu su juo dirbti vėliau. Tam naudosime <b>kintamuosius</b>.
Sukurkime kintamąjį pavadinimu <b>suma</b> ir priskirkime jam reikšmę, kurią gauname pridėję 7 prie 1.
Tada išspausdiname kintamajam priskirtą reikšmę.

```python
# Sukuriame kintamąjį pavadinimu suma ir suteikite jam reikšmę 7 + 1
suma = 7 + 1

# išspaudiname kintamojo suma reikšmę
print(suma)
```

Apskritai, norėdami dirbti su kintamaisiais, pirmiausia turite parinkti tinkamą jiems vardą. Idealiu atveju kintamųjų pavadinimai yra trumpi ir aiškiai juos apibūdinantys. Jie taip pat turi atitikti keletą reikalavimų:
 - Juose negali būti tarpų (pvz., testinis kintamasis yra neleistina forma)
 - Juose gali būti tik raidžių, skaičių ir apatinių brūkšnių jokių kitokių symboliu (pvz., test_kint! neleistina forma)
 - Jie turi prasidėti raide arba pabraukimu (pvz., 1_var neleistina forma)

Sukurdami kintamąjį, turite naudoti lygybės ženklą =, tam kad priskirtumėte jam norimą reikšmę.
Visada galite pažvelgti į kintamojo priskirtą reikšmę naudodami print() ir skliausteliuose įrašydami kintamojo pavadinimą.
Laikui bėgant išmoksite pasirinkti tinkamus Python kintamųjų pavadinimus. Viskas gerai jeigu dabar jaučiates nepatogiai, o geriausias būdas mokytis yra tiesiog išmgėginti daug python programavimo.
______________________

Visada galite pakeisti kintamajam priskirtą reikšmę, nepaisydami ankstesnės reikšmės.
Žemiau esančiame kodo langelyje keičiame mano_kin reikšmę nuo 1 iki 100.

```python
# Nustatykite naujo kintamojo reikšmę į 1
mano_kin = 1

# Išspausdinkite mano_kin priskirtą reikšmę
print(mano_kin)

# Pakeiskite kintamojo reikšmę į 100
mano_kin = 100

# Išspausdinkite naują reikšmę, priskirtą my_var
print(mano_kin)
```

Atkreipkite dėmesį, kad paprastai, kai apibrėžiate kintamąjį kodo langelyje, visi tolesni kodo langeliai taip pat turi prieigą prie šio kintamojo. Pavyzdžiui, mes naudojame kitą kodo langelį, kad pasiektume my_kim (iš aukščiau esančio kodo langelio) ir suma (iš dar ankstesnio kodo langelio) reikšmes.

```python
print(suma)
print(mano_kin)
```

Kitas kodo langelis nurodo Python padidinti dabartinę mano_kin reikšmę.
Norėdami tai padaryti, vis tiek turime naudoti my_kin = kaip anksčiau. Taip pat, kaip ir anksčiau, nauja reikšmė, kurią norime priskirti kintamajam, yra dešinėje nuo lygybės ženklo.

```python
mano_kin = mano_kin + 5
print(mano_kin)
```

Galima naudojanti kelis kintamuosius iškarto tai yra įprasta praktika.
Kitame kodo langelyje apskaičiuojame sekundžių skaičių per ketverius metus. Šiam skaičiavimui naudojami penki įvesties duomenys.

```python
# Kintamieji
met = 4
die = 365 
val = 24
min = 60
sec = 60

# Suskaičiuokime kiek sekundžių yra keturiuose metuose
viso = sec * min * val * die * met
print(viso)
```

Kaip apskaičiuota aukščiau, per ketverius metus yra 126144000 sekundžių.

Atkreipkite dėmesį, kad šį skaičiavimą galima atlikti be kintamųjų, tik su reikšmėmis 60 * 60 * 24 * 365 * 4, tačiau daug sunkiau patikrinti, ar skaičiavimas be kintamųjų neturi klaidų, nes jis nėra taip gerai įskaitomas. Kai naudojame kintamuosius (pvz., metų skaičius, dienų_per metus ir kt.), galime geriau sekti kiekvieną skaičiavimo dalį ir lengviau patikrinti bei ištaisyti klaidas.

Atkreipkite dėmesį, kad ypač naudinga naudoti kintamuosius, kai gali keistis įvesties reikšmės. Pavyzdžiui, tarkime, kad norime šiek tiek pakeisti savo skaičiavimą, atnaujindami dienų skaičių per metus nuo 365 iki 365,25, kad būtų atsižvelgta į keliamuosius metus. Tada galime pakeisti dienų per metus kintamojo priskirtą reikšmę nekeisdami jokių kitų kintamųjų ir atlikti skaičiavimą iš naujo.

```python
die = 365.25
naujas_viso = sec * min * val * die * met
print(naujas_viso)
```

## 3.Funkcijos

Šioje dalyje sužinosite, kaip rašyti kodą su funkcijomis. Funkcija yra kodo blokas, skirtas konkrečiai užduočiai atlikti. Kaip matysite, funkcijos leis jums atlikti maždaug tą patį skaičiavimą kelis kartus nedubliuojant tų pačių komandų. Pradedame nuo paprastos funkcijos pavyzdžio. Funkcija `pridek_tris()` priima bet kokį skaičių, prideda prie jo 3 ir grąžina rezultatą.

```python
def pridek_tris(ivestis):
    isvestis = ivestis + 3
    return isvestis
```

Kiekviena funkcija susideda iš dviejų dalių: antraštės ir kūnas. Funkcijos antraštė apibrėžia funkcijos pavadinimą ir jos argumentą (-us).<br>
<code>**Antraštė**</code> prasideda <code>**def**</code>, kuri praneša python, kad ketiname apibrėžti funkciją.

 - Pavyzdyje funkcijos pavadinimas yra `pridek_tris`
 - Pavyzdyje argumentas yra `ivestis`. Argumentas yra kintamojo, kuris bus naudojamas kaip funkcijos įvestis, pavadinimas. Jis visada pateikiamas skliausteliuose, kurie atsiranda iškart po funkcijos pavadinimo. (Atkreipkite dėmesį, kad funkcija taip pat gali neturėti argumentų arba turėti kelis argumentus. Keletą pavyzdžių pamatysite vėliau)
 - Kiekvienoje funkcijoje už skliaustelių, kuriuose yra funkcijos argumentas (-ai), turi būti dvitaškis :
 
 Funkcijos <code>**kūnas**</code> apibrėžia darbą, kurį atliks funkcija.
  - Kiekviena kodo eilutė funkcijos kune turi būti atitraukta tiksliai keturiais tarpais. Tai galite padaryti keturis kartus paspausdami tarpo klavišą arba vieną kartą paspaudę klaviatūros mygtuką „Tab“. (Kai kada kodą gali tekti įtraukti daugiau nei keturiais tarpais, bet daugiau apie tai sužinosite vėliau )
  - Funkcija atlieka savo darbą vykdydama visas eilutes iš viršaus į apačią.
  
Aukščiau esantis kodo langelis tik apibrėžia funkciją `pridek_tris`, bet jos nevykdo. Funkcijos kūno detalės bus aiškesnės kai iš tikrųjų įvykdysime šią funkciją.
Kai vykdome funkciją, taip taip pat gali būti vadinama funkcijos iškvietimas.
Žemiau esančiame kodo langelyje paleidžiame funkciją su 2 kaip įvesties reikšmę. Mes apibrėžiame naują kintamąjį <code>**naujas_nr**</code>, kuris nustatomas kaip funkcijos išvestis.

```python
# Paleiskite funkciją naudodami 5 kaip įvestį
naujas_nr = pridek_tris(5)

# patikriname ar reikšmė yra 8 kaip ir tikimasi
print(naujas_nr)
```

Aukščiau pateiktame pavyzdyje už jus buvo pasirinktas funkcijos pavadinimas. Įvardindami savo funkcijas, naudokite tik mažąsias raides, žodžius atskirdami apatiniais brūkšniais, o ne tarpais. Funkcijų įvardijimas laikui bėgant jausis natūraliai, o pagridininė taisyklė pavadinti funkcijai taip kad jos pavadinimas atsakytu į klausimą, - ką ši funkcija daro ?

Dabar kai jau žinote šiek pagrindus, galime pereiti prie pavyzdžio su ilgesniu skaičiavimu atlyginimo mokesčiams. Ir pamėginkime taikyti anglų kalbą įvardinant funkciją ir kintamuosiuose, tai irgi yra industrijos strandartas visame pasaulyje.
Situacija:
 - Mokesčių tarifas yra 12 % (kitaip tariant, 12 % atlyginimo yra paimama mokesčiams, o parsineša gauni likusią 88 % dalį)
 - Valandinio darbo tarifas yra 60 €

```python
def get_payment(hours_worked):
    bruto = hours_worked * 60
    neto = bruto * (1 - .12)
    return neto
```

```python
payment = get_payment(40)
print(payment)
```

Aukščiau esanti funkcija apskaičiuoja atlyginimą pagal dirbtų valandų skaičių. 
Funkcija yra šiek tiek sudėtingesnė nei pirmajame pavyzdyje, funkcija turi vieną argumentą (darbo valandų skaičius). Funkcijų kune mes:
 - Naudojame `hours_worked` argumento reikšmę, kad nurodytumėte naujo kintamojo `bruto` reikšmę po aritemtinių skaičiavimų operacijos.
 - Panaudojame `bruto` reikšmę, kad nurodytumėte naujo kintamojo `neto` reikšmę.

## 4.Duomenų typai

Kai kuriate kintamąjį jis turi reikšmę su atitinkamu duomenų tipu. Yra daug skirtingų duomenų tipų, pvz., <code>**sveikieji skaičiai (integer), trupmeninės reikšmės (float), loginės reikšmės (booleans) ir eilutės (strings)**</code>, netrukus juos visus aptarsime. Bet tai tik nedidelė ištrauka iš visų galimų duomenų tipų – taip pat yra <code>**žodynai (dictionaries), rinkiniai (sets), sąrašai (list), kortelės (tuple)**</code>.<br>
Duomenų tipai yra svarbūs, nes jie nustato, kokius veiksmus su jais galite atlikti. Pavyzdžiui, galite padalyti dvi trupmenas (floats) ar skaičius (integers), bet negalite padalinti dviejų eilučių (strings). Pavyzdžiui, 8.5 / 2.0 yra logiška ir prasminga, bet "katė" / "šuo" turbūt nelabai. Kad išvengtume tokių klaidų, operacijos turi atitinka duomenų tipus.

## 5.Sąlygos ir sąlyginės išraiškos

Dirbdam su funkcijomis matėme kad pakeitę įvesties reikšmę, gaunate skirtingą išvestį. Pavyzdžiui, funkciją `pridek_tris()`, kuri tiesiog prideda tris prie bet kurio skaičiaus ir grąžina rezultatą. Tada pridek_tris(7) pateiks išvestį 10 (= 7 + 3), o pridek_tris(8) – 11 (= 8 + 3). Atkreipkite dėmesį, kad nesvarbu, kokia įvestis, veiksmas, kurį atlieka funkcija, visada yra tas pats: ji visada prideda tris.

<code>**Tačiau jums gali prireikti funkcijos, kuri atlieka veiksmą, priklausantį nuo įvesties reikšmės!**</code>. Pavyzdžiui, jums gali prireikti funkcijos pridek_tris_ar_penkis(), kuri prideda tris, jei įvestis yra mažesnė nei 10, ir prideda penkis, jei įvestis yra 10 ar daugiau. Tada pridek_tris_ar_penkis(1) pateiks 4 (= 1+3), bet pridek_tris_ar_penkis(11) pateiks 16 (= 1 1+ 5). Šiuo atveju veiksmas, kurį atlieka funkcija, priklauso nuo įvesties. Reikia panaudoti sąlygas ir sąlyginius teiginius, kad pakeistumėte funkcijos vykdymo eigą.

### Sąlygos

Programavime sąlygos yra teiginiai, kurie yra teisingi arba klaidingi. Yra daug skirtingų būdų, kaip rašyti sąlygas python kalboje, tačiau kai kurie dažniausiai pasitaikantys sąlygų rašymo būdai yra kai lygina dvi skirtingas reikšmes. Pavyzdžiui, galite patikrinti, ar 1 yra didesnis nei 3.

```python
print( 1 > 3)
```

Python tai identifikuoja kaip klaidingą, nes 1 nėra didesnis nei 3. Taip pat galite naudoti sąlygas, kad palygintumėte kintamųjų reikšmes. Kitame kodo langelyje a_kin reikšmė yra 1, o b_kin - 2. Patikriname ar a_kin yra mažesnis nei 1 (tai yra klaidinga), ir patikriname, ar b_kin yra didesnis arba lygus a_kin (tai yra tiesa).

```python
a_kin = 1
b_kin = 2

print(a_kin < 1)
print(b_kin >= a_kin)
```

```python
print(a_kin != b_kin)
```

Įprastų simbolių, kuriuos galite naudoti tikrinti sąlygoms, sąrašą rasite toliau pateiktoje žemiau lentelėje.

| Simbolos  | Reikšmė           |
|         :-|                :- |
|  ==       | lygu              |
|  !=       | nelygu            |
|  <        | mažiau nei        |
|  <=       | mažiau arba lygu  |
|  >        | daugiau           |
|  =>       | daugiau arba lygu |


<code>**Svarbi pastaba**</code> kai tikrinate, ar dvi reikšmės yra lygios, įsitikinkite, kad naudojate ženklą ==, o ne ženklą = , nes = yra reikšmės priskirimo ženklas<br>
 - a_kin == 1, tikrina ar kintamojo a_kin reikšmė yra 1
 - a_kin =  1, nustato kintamojo a_kin reikšmę būti 1.

### Sąlyginiai teiginiai

Sąlyginiai teiginiai naudoja sąlygas, kad pakeistų jūsų funkcijos veikimą. Jie patikrina sąlygos reikšmę ir, jei sąlyga įvertinama kaip True, tada vykdomas tam tikras kodo blokas. (Priešingu atveju, jei sąlyga yra klaidinga, kodas nepaleidžiamas.)

### if teiginiai

Paprasčiausias sąlyginio teiginio tipas yra `if` teiginys. To pavyzdį galite pamatyti žemiau esančioje funkcijoje measure_temp(). Funkcija priima kūno temperatūrą (celsijaus laipsniais) kaip įvestį.

```python
def measure_temp(temp):
    # nustatome pradinį pranešimą
    message = "normali temperatūra!"
    # Atnaujinkite pranešimo reikšmę, tik jei temperatūra aukštesnė nei 37
    if temp > 37:
        message = "karščiuojate!"
    return message
```

Funkcijos kūna sudaro šie operacijų blokai:
- Iš pradžių pranešimas nustatomas į "normali temperatūra".
- Tada, jei kintamojo temp reiškmė didesnė nei 37 , sąlyga if temp > 37 yra tiesa (True) (pvz., kūno temperatūra yra didesnė nei 37 °C), pranešimas atnaujinamas į "karščiuojate!". Kitu atveju (mažesnės temperaturos atveju), salyga temp > 37  yra False, pranešimas neatnaujinamas.
- Galiausiai funkcija grąžina pranešimą.

Kitame kodo langelyje kviečimame funkciją ir jai duodame temperatūros kintamajo reiškmę kuri vus 36°C. Pranešimas bus "normali temperatūra", nes šiuo atveju temperatūra yra žemesnė nei 387°C (temp > 37 sąlyga reiškia False).

```python
print(measure_temp(36))
```

```python
print(measure_temp(39))
```

<code>**Atkreipkite dėmesį, kad yra du įtraukos lygiai:**</code>
 - Pirmasis įtraukos lygis yra todėl, kad visada turime įtraukti kodo operacijų blokus funkcijos viduje.
 - Antrasis įtraukos lygis yra todėl, kad taip pat turime įtraukti kodo bloką, priklausantį sąlygos teiginiui "if". (Taip pat turėsime įtraukti kodų blokus "elif" ir "else" teiginiams.)

<code>**Pastaba**</code>, kad grąžinimo operacija `return` nėra įtrauktas po sąlygos teiginiu "if", todėl jis visada yra įvykdomas, nesvarbu, ar temp > 37 yra teisinga, ar klaidinga !

### if else

Mes galime naudoti teiginius "else" norėdami paleisti kodą, jei teiginys "if" yra klaidingas. Kodo blokas po sakiniu "if" paleidžiamas, jei teiginys yra tiesa True, o kodo blokas "else" paleidžiamas, jei teiginys yra klaidingas.

```python
def measure_temp_with_else(temp):
    if temp > 37:
        message = "karščiuojate!"
    else:
        message = "normali temperatūra"
    return message
```

```python
measure_temp_with_else(38)
```

Šios funkcijos `measure_temp_with_else()` elgesys yra lygiavertis funkcijai  `measure_temp()`. 
Kitame kodo langelyje kviečiame šią funkciją ir duodama temperatūra yra 36°C. Šiuo atveju temp > 37 įvertinama kaip False, todėl vykdomas kodas, esantis sakinyje "else", ir pranešimas "normali temperatūra" yra grąžinamas.

```python
print(measure_temp_with_else(36))
```

Kaip ir ankstesnėje funkcijoje, kodo blokus atitraukiame po teiginių "if" ir "else". Tai labai svabu!

### if, elif, else

Galime naudoti `elif` (kuris yra `else if` trumpinys), kad patikrintume, ar kelios sąlygos gali būti teisingos. Funkcija žemiau:

```python
def measure_temp_with_elif(temp):
    if temp > 37:
        message = "karščiuojate!"
    elif temp > 35:
        message = "normali temperatūra"
    else:
        message = "per žema temperatūra!"
    return message
```

- Pirmiausia patikrinama, ar temp > 37. Jei tai tiesa, tada pranešimas nustatomas į "karščiuojate!".
- Kol pranešimas dar nenustatytas, funkcija patikrina, ar temp > 35. Jei tai tiesa, pranešimas nustatomas į "normali temperatūra".
- Tada, jei vis dar nenustatytas joks pranešimas, teiginys `else` užtikrina, kad pranešimas būtų nustatytas į "per žema temperatūra!". 
- Galiausia pranešimas yra gražinamas.

Galite galvoti apie `elif` kaip sakant, - gerai, ta ankstesnė sąlyga (pvz., temp > 37) buvo klaidinga, todėl patikrinkime, ar ta kita sąlyga (pvz., temp > 35) gali būti teisinga.

Žemiau esančiame kodo langelyje kviečiame funkciją su 36 °C. Operacijos vyks `elif` kodo bloke, nes temp > 37 bus klaidinga, o temp > 35 bus teisinga. Kai šis kodas paleidžiamas, funkcija tokiomis sąlygomis visiškai praleis teiginį `else` ir grąžins pranešimą.

```python
print(measure_temp_with_elif(33))
```

Galiausiai išbandome atvejį, kai temperatūra žemesnė nei 35°C. Kadangi sąlyginiai sakiniai `if` ir `elif` yra vertinami kaip False, bus vykdomas kodo blokas `else`.

```python
print(measure_temp_with_elif(34))
```

### Kalkuliacijos

Iki šiol pateiktuose pavyzdžiuose buvo naudojami sąlyginiai teiginiai sprendžiant, kaip nustatyti kintamųjų reikšmes.
Tačiau taip pat galite naudoti sąlyginius teiginius, kad atliktumėte įvairius skaičiavimus. Kitame pavyzdyje tarkime, kad gyvenate šalyje, kurioje yra tik dvi mokesčių grupės. Kiekvienas, uždirbantis mažiau nei 12000, moka 25 % mokesčių, o kas uždirba 12000 ar daugiau – 30 %. Toliau pateikta funkcija apskaičiuoja, kiek mokesčių reikia jums mokėti.

```python
def get_taxes(earnings):
    if earnings < 12000:
        tax = .25 * earnings
    else:
        tax = .30 * earnings
    return tax
```

```python
a_person_taxes = get_taxes(15000)
b_person_taxes = get_taxes(9000)

print(a_person_taxes)
print(b_person_taxes)
```

Kiekvienu atveju mes iškviečiame funkciją `get_taxes()` ir naudojame grąžintą reikšmę kintamojo reikšmei nustatyti.
 - a_person_taxes apskaičiuojame mokesčius, kuriuos turi mokėti šis asmuo, uždirbantis 15 tūkst. Šiuo atveju iškviečiame funkciją get_taxes(), kai pajamos yra 150000. Taigi, pajamos < 12000 yra True, o tax nustatytas į 0.25 * 9000. Tada grąžiname mokesčių vertę.
 - b_person_taxes apskaičiuojame mokesčius, kuriuos turi mokėti asmuo, uždirbantis 9 tūkst. Šiuo atveju iškviečiame funkciją get_taxes() su uždarbiu, nustatyta į 90000. Taigi uždarbis < 12000 yra False, progra o tax nustatytas į 0,30 * 15000. Tada grąžiname tax vertę.

### Keli elif teiginiai

Iki šiol matėte kaip funkcijoje `elif` naudojamas tik vieną kartą. Tačiau `elif` teiginių skaičius neribojamas. Pavyzdžiui, kitame kodo bloke apskaičiuojama vaisto dozė (mililitrais), kurią reikia duoti vaikui, atsižvelgiant į jo svorį (kilogramais).

```python
def get_dose(weight):
    # Visiems, sveriantiems mažiau nei 5,2 kg, dozė yra 1,25 ml
    if weight < 5.2:
        dose = 1.25
    elif weight < 7.9:
        dose = 2.5
    elif weight < 10.4:
        dose = 3.75
    elif weight < 15.9:
        dose = 5
    elif weight < 21.2:
        dose = 7.5
    # Dozavimas yra 10 ml sveriantiems 21,2 kg ar daugiau
    else:
        dose = 10
    return dose
```

```python
print(get_dose(22))
```

- Šiuo atveju `if` buvo False, o visi `elif` teiginiai vertinami kaip klaidingi, kol prieiname prie sąlygos kur svoris < 15,9, o tada tai yra True, dozė nustatoma į 5.
- Kai vienas teiginys `elif` įvertinamas kaip True yra paleidžiamas tas kodo blokas, funkcija praleidžia visus likusius `elif` ir `else` teiginius. Praleidus juos, belieka grąžinimas `return`, kuris grąžina dozės reikšmę.
- Čia svarbi  yra `elif` teiginių tvarka! Kitaip išdėstę pareiškimus gausite skirtingą rezultatą.

## 6.Sąrašas (masyvas)

Atliekant duomenų mokslą, reikia būdo kažkaip tvarkyti duomenyss, kad galėtumėte efektyviai su jais dirbti. Python kabla turi daug duomenų struktūrų, skirtų jūsų duomenims saugoti, pvz., sąrašus, rinkinius, žodynus ir eilutes. Šioje dalyje sužinosite, <code>**kaip dirbti su Python sąrašais (list).**</code><br>
Tarkime jūsų uždavinio tikslas yra suskirstyti gėlės rūšis tik pagal jos išvaizda (nuotrauka). Tai yra gana įprasta kompiuterinės regos užduotis ir ji vadinama vaizdo klasifikavimu. Siekdami šio tikslo, tarkime, kad duomenyse pirmą sutvarkote gėlių rūšių pavadinimus. Vienas iš būdų tai padaryti yra sutvarkyti pavadinimus Python eilutėje (string).

```python
flowers = "pink primrose, hard-leaved pocket orchid,canterbury bells,sweet pea,english marigold,tiger lily,moon orchid,bird of paradise,monkshood,globe thistle"

print(type(flowers))
print(flowers)
```

Bet dar geriau yra pateikti tuos pačius duomenis sąraše. Norėdami sukurti tokį sąrašą, turite naudoti laužtinius skliaustus `[]` ir atskirti kiekvieną elementą sąraše kableliu. Kiekvienas šio sąrašo elementas yra  eilutė (string), todėl kiekvienas įrašas yra kabutėse.

```python
a = []
```

```python
print(type(a))
```

```python
flowers_list = [
    "pink primrose", 
    "hard-leaved pocket orchid", 
    "canterbury bells", 
    "sweet pea",
    "english marigold", 
    "tiger lily", 
    "moon orchid", 
    "bird of paradise", 
    "monkshood",
    "globe thistle"
]
```

```python
flowers_list
```

```python
print(type(flowers_list))
print(flowers_list)
```

Iš pirmo žvilgsnio gal tai neatrodo per daug skirtingai, ar pateikiate informaciją  eilutėje (string) ar sąraše (list). Tačiau, kaip pamatysite, yra daug užduočių, kurias galite daug lengviau atlikti naudodami sąrašą. Pavyzdžiui, sąrašas palengvins:
 - gauti įrašą nurodytoje vietoje (pirmoje, antroje, trečioje ir tt.).
 - patikrinkite įrašų skaičių.
 - pridėti ir pašalinti įrašus sąraše.

### Ilgis

Įrašų skaičių bet kuriame sąraše galime suskaičiuoti naudodami python in-built funkcija `len()`. Kviečiat šią funkcija skliausteliuose reikia nurodyti tik sąrašo pavadinimą.

```python
print( len(flowers_list) )
```

### Indeksavimas

Mes galime nurodyti bet kurį sąrašo elementą pagal jo vietą sąraše (pirma, antra, trečia ir tt). Tai vadinama indeksavimu. Atminkite, kad Python naudoja <code>**nulinį indeksavimą**</code>, o tai reiškia, kad:
 - Norėdami ištraukti pirmąjį sąrašo įrašą, naudokite indeksą 0,
 - Norėdami ištraukti antrąjį sąrašo įrašą, naudokite 1
 - Kad ištraukti galutinį sąrašo įrašą, naudokite vienu mažiau nei sąrašo ilgis.

```python
print("Pirmas įrašas:", flowers_list[0])
print("Antras įrašas:", flowers_list[1])

# Sąrašo ilgis yra 10, todėl mes nurodome 9
print("Paskutinis įrašas:", flowers_list[9])
```

<code>**Pastaba:**</code> Galbūt pastebėjote, kad aukščiau esančiame kodo langelyje mes naudojame vieną print() norėdami spausdinti kelis elementus (tiek Python eilutę (pvz., "Pirmas įrašas:"), tiek reikšmę iš sąrašo (pvz., flowers_list[0]) ). Norėdami spausdinti kelis dalykus Python naudodami vieną komandą, tereikia juos atskirti kableliais.

### Pjaustymas

Taip pat galite ištraukti sąrašo segmentą (pavyzdžiui, pirmus tris arba paskutinius du įrašus). Tai vadinama pjaustymu (slicing). Pavyzdžiui:
 - Norėdami ištraukti pirmuosius x įrašus, naudokite [:x];
 - Kad ištraukti paskutinius y įrašus, naudokite [-y:].

```python
a_list = [ 1, 2, 3, 4, 5]
```

```python
a_list[:3]
```

```python
a_list[-2:]
```

```python
print("Pirmieji trys įrašai:", flowers_list[:3])
print("Paskutiniai du įrašai:", flowers_list[-2:])
```

Kaip matote aukščiau, kai supjaustome sąrašą, jis grąžina naują sutrumpintą sąrašą su pjausnio reikšmėmis.

### Elementų pašalinimas

Pašalinkite elementą iš sąrašo naudodami `.remove()` sąrašo metodą, o norimą pašalinti elementą įrašykite skliausteliuose. <code>**Pastaba:**</code> metodas yra klasės funkcija, šiuo atveju klasė yra sąrašas (list)

```python
a_list = [1, 1, 1, 5, 5]
```

```python
a_list
```

```python
a_list.append(6)
```

```python
a_list
```

```python
flowers_list.remove("globe thistle")
print(flowers_list)
```

### Elementų pridėjimas

Pridėkite elementą į sąrašą naudodami `.append()` metodą, o norimą įtraukti elementą įrašykite skliausteliuose.

```python
flowers_list.append("daisies")
print(flowers_list)
```

### Sąrašai skirti ne tik eilutėms

Iki šiol dirbome tik su sąrašais, kuriuose kiekvienas sąrašo elementas yra eilutė (string). Tačiau sąrašuose gali būti elementų su bet kokiu duomenų tipu, įskaitant loginius, sveikųjų ir trupmeninių skaičių duomenis.
Pavyzdžiui, apsvarstykite knygų išpardavimą kietais viršeliais pirmąją 2000 m. balandžio mėnesio savaitę mažmeninės prekybos parduotuvėje.

```python
book_sales = [139, 128, 172, 139, 191, 168, 170]
```

Čia sales yra sveikųjų skaičių sąrašas. Panašiai kaip dirbdami su eilutėmis, vis tiek galite atlikti tokius veiksmus, kaip nustatyti ilgį, ištraukti atskirus įrašus ir išplėsti šį sąrašą.

```python
print("Length of the list:", len(book_sales))
print("Entry at index 2:", book_sales[2])
```

Taip pat galite gauti mažiausią reikšmę su `min()` ir didžiausią reikšmę su `max()` python in-built funkcijomis. Norėdami susumuoti kiekvieną sąrašo elemento reikšmę, naudokite `sum()`.

```python
#print("Mažiausiai:", min(book_sales))
print("Daugiausiai:", max(book_sales))
print("Viso parduota:", sum(book_sales))
```

Panašiaus skaičiavimus galime atlikti ir su sąrašo dalimis. Kitame kodo langelyje paimame pirmųjų penkių dienų sumą (suma(book_sales[:5])), tada padaliname iš penkių, kad gautume vidutinį parduotų knygų skaičių per pirmąsias penkias dienas.

```python
sum(book_sales[:5])/5
```
